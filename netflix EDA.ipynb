```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
print(netflix_df.columns)
print(netflix_df['release_year'])
print(netflix_df['duration'])
```

**_What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration._**

```python
#Filter the DataFrame to include only movies from the 1990s
released_1990 = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
```

```python
#  Find the most frequent movie duration in the 1990s
most_frequent_duration = released_1990['duration'].mode()[0]

#Save the result as an integer called duration
duration = int(most_frequent_duration)

#Plotting histogram
plt.hist(released_1990['duration'])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel(f'Duration (minutes)')
plt.ylabel('Number of movies')
plt.show()

print(f"The most frequent movie duration in the 1990s is {duration} minutes.")
```

**_A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count_**

```python
# filter to include only action movies
action_movies_1990s = released_1990[(released_1990['genre'] == 'Action')]

# Filter to include only movies with duration less than 90 minutes
short_action_movies_1990s = action_movies_1990s[action_movies_1990s['duration'] < 90]

# Count the number of short action movies
short_movie_count = short_action_movies_1990s.shape[0]

print(f"The number of short action movies released in the 1990s is {short_movie_count}.")
```
